```python
import pandas as pd
import re
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score
```

```python
import nltk
nltk.download('punkt')
nltk.download('stopwords')
```

```python
od.download('https://www.kaggle.com/datasets/jainpooja/fake-news-detection')
```

```python
true_data = pd.read_csv("/content/fake-news-detection/True.csv")
Fake_data = pd.read_csv("/content/fake-news-detection/Fake.csv")
```

```python
true_data.head()
```

```python
true_data['label'] = 1
Fake_data['label'] = 0
```

```python
true_data.shape
```

```python
Fake_data.shape
```

```python
combined_data = pd.concat([true_data, Fake_data])
```

```python
data = combined_data.sample(frac=1).reset_index(drop=True)
```

```python
data
```

```python
data = data.drop(['title','subject','date'], axis = 1)
```

```python
data
```

```python
def preprocess_text(text):
    # Convert to lowercase
    text = text.lower()

    # Remove special characters, numbers, and punctuations
    text = re.sub(r'[^a-zA-Z\s]', '', text)

    # Tokenize the text
    tokens = word_tokenize(text)

    # Remove stop words
    stop_words = set(stopwords.words('english'))
    tokens = [word for word in tokens if word not in stop_words]

    # Join tokens back into a string
    processed_text = ' '.join(tokens)

    return processed_text
```

```python
data.text = data.text.apply(preprocess_text)
```

```python
X = data.text
Y = data.label
```

```python
x_train, x_test, y_train, y_test = train_test_split(X , Y ,test_size = 0.2)
```

```python
vector = TfidfVectorizer()
train = vector.fit_transform(x_train)
test = vector.transform(x_test)
```

```python
model = DecisionTreeClassifier()
model.fit(train,y_train)
```

```python
pred = model.predict(test)
```

```python
print("The accuracy is = ",accuracy_score(pred,y_test))
```
